```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def weighted_avg(x, weight, factor):    
    with warnings.catch_warnings():        
        warnings.simplefilter('ignore')        
        tmp = x[[weight, factor]].dropna()        
        weighted_sum = (tmp[weight] * tmp[factor]).sum()        
        count_sum = tmp[weight].sum()        
        return weighted_sum / count_sum
```

```python
df = pd.read_csv('data/smolt_dataset_transfers_until2023Feb28_narrow.csv')
df2 = pd.read_csv('data/sw_locus_fw_locus_population_with_counts.csv')
density = pd.read_csv('data/locus_density.csv', encoding='windows-1252')
targets = pd.read_csv('data/targets.csv', encoding='windows-1252')

df2['transfer_date'] = pd.to_datetime(df2['transfer_date'])
df2['transfer_year']=df2['transfer_date'].dt.year

key_columns = ['locus_id', 'fish_group_id', 'transfer_year']
```

```python
df2=df2.merge(density, left_on='from_locus_id',right_on='locus_id',how='left')
df2['biomass'] = df2['from_avg_weight']/1000*df2['from_count_stocking']
df2['density']=df2['biomass']/df2['volume']
```

```python
df['volume'].describe()
```

```python
df_density_grouped=pd.DataFrame(df2.groupby(['to_locus_id', 'to_fish_group_id', 'transfer_year']).apply(weighted_avg, 'from_count_stocking', 'density'))
df_density_grouped.columns=['density_FW']
```

```python
df=df.merge(density, left_on='locus_id',right_on='locus_id',how='left')
```

```python
df['biomass'] = df['to_avg_weight']/1000*df['to_count_stocking']
df['density_SW']=df['biomass']/df['volume']
```
